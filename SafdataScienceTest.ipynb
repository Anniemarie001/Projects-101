```python
#filtering out duplicated rows
biopics.sort_values("title", inplace = True)

biopics.drop_duplicates(subset ="title",
                     keep = False, inplace = True)
biopics
```

```python
#renaming  the variable box_office to earnings
biopics_df= biopics.rename(columns={'box_office': 'earnings'})
biopics_df
```

```python
#filter out rows for which earnings are missing
biopics_df.dropna(subset=['earnings'],inplace=True)

biopics_df
```

```python
#keep only movies  released in year 1990 or later
biopics_df= biopics_df[biopics_df['year_release']>=1990]
#riders = riders[riders['active_rider_age'] <70]
biopics_df['year_release'].value_counts()
```

```python
#creating a new variable called lead_actor_actress_known that is False if lead_actor_actress is NaN and True of otherwise
def f(row):
    if row['lead_actor_actress'] == "NaN":
        lead_actor_actress_known = False
    else:
         lead_actor_actress_known = True
    return  lead_actor_actress_known
```

```python
biopics_df['lead_actor_actress_known'] = biopics_df.apply(f, axis=1)
biopics_df
```

```python
#reordering the columns 
biopics_new=biopics_df[['title','year_release','earnings','country','type_of_subject',
                        'lead_actor_actress','lead_actor_actress_known']]
biopics_new
```

```python
#sort the rows in descending order by earnings
biopics_sorted = biopics_new.sort_values(by ='earnings', ascending = 0)
#print("Sorting rows by earnings:\n \n", biopics_sorted)
biopics_sorted
```
