```python
#first we import pandas
import pandas as pd
#we can check the version in use as well
print(pd.__version__)
```

```python
#creating data frames
#Example 1
Data={"Fruits":["Apple", "Mangoes", "Bananas", "Oranges","Grapes"],
    "Prices":[35,30,20,15,100]
     }#our data
#loading the data in a data frame
df=pd.DataFrame(Data)
#printing the data
print(df)

```

```python
#Example 2
students={"Names":["Ann","Joy","Kamau","Moureen"],
         "Age":[21,22,24,23]}
#loading the data in a dataframe
students_df=pd.DataFrame(students)
#print it
print(students_df)
```

```python
#locating specific rows
print(df.loc[1])
print(students_df.loc[3])
```

```python
#SERIES CREATION
#Series is a type of list in pandas which can take integer values, string values, double values and more 
# Series  can only contain single list with index
a=[2,3,"ann",2.3]
myvar=pd.Series(a)
print(myvar)

```

```python
#Creating labels or indexing
a=[1,2,3]
myvar=pd.Series(a,index=["X","Y","Z"])
print(myvar)
```

```python
#indexing students
Students=["ANN","JOY","JOHN","MOUREEN"]
students_df=pd.Series(Students, index=["001","002","003","004"])
print(students_df)
```

```python
#KEYS\VALUES OBJECT AS SERIES
#EXAMPLE1
# using all keys to create pandas series dataset
Months={"Month1":"January", "Month2": "Feb","Month3":"March"}
month_df=pd.Series(Months)
print(month_df)
```

```python
# using some keys to create pandas series dataset
Months={"Month1":"January", "Month2": "Feb","Month3":"March"}
month_df=pd.Series(Months, index=["Month1","Month2"])
print(month_df)
```

```python
#HANDLING MISSING DATA
#Pandas treats None and NaN as essentially interchangeable for indicating missing or null values.

#Pandas Methods for missing values:

 #isnull(): Generate a boolean mask indicating missing values
 #notnull(): Opposite of isnull()
 #dropna(): Return a filtered version of the data
 #fillna(): Return a copy of the data with missing values filled or imputed
```

```python
#lets check if we have missing values in our data
data.isnull().sum().any()
#False means there are no missing values
```

```python
data.notnull().sum().any()
```

```python
#check missing values in columns
data.isnull().sum()
```

```python
#now incase we found  missing values how do we drop them?
#we use dopna() which drops all rows with missing values
clean_data=data.dropna()
data.isnull().sum().any()

```

```python
#not all the time we drop rows with missing values. Sometimes we can fill them with valid value like the mean.
#or This value might be a single number like zero, or it might be some sort of imputation or interpolation from the good values
#FILLING THE MISSING VALUES

data['FEMALE'].fillna(data['FEMALE'].mean(),inplace=True)
data["MALE"].fillna(data["MALE"].mean(), inplace=True)
```
